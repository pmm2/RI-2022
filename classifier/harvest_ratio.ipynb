```python
files_list = [
    # "www.shopb.com.br.txt",
    # "www.mercadolivre.com.br.txt",
    "www.bigboygames.com.br.txt",
    "www.ibyte.com.br.txt",
    "www.shockgames.com.br.txt",
    "www.vnsgames.com.br.txt"
]

url_list = []

for file_name in files_list:
    file = open(f"./harvested_links/{file_name}")
    lines = file.readlines()
    file.close

    for url in lines:
        if len(url) > 1:
            url_list.append(url.strip("\n"))

total_urls = len(url_list)
print("Total de urls:", total_urls)
```

```python
import requests
from bs4 import BeautifulSoup

contents = []

# Function to remove tags
def remove_tags(html):
    soup = BeautifulSoup(html, "html.parser")
  
    for data in soup(['style', 'script']):
        data.decompose()
  
    return ' '.join(soup.stripped_strings)

for url in url_list:
    r = requests.get(url)
    filtered_content = remove_tags(r.content)
    contents.append(filtered_content)
    print(url)
```

Loading model

```python
from joblib import load

classifier_model = load('./persistence/classifier_model.joblib')
```

```python
len(contents)
```

```python
predictions = classifier_model.predict(contents)
```

```python
predictions
```

```python
absolute_harvest = 0
for prediction in predictions:
    if prediction:
        absolute_harvest += 1

print("Predições positivas absolutas:", absolute_harvest)

harvest_ration = 100 * (absolute_harvest / total_urls)

print(f"Harvest Ration: {harvest_ration}%")
```
